```python
df1=df1[~df1['rate (out of 5)'].isna()]
```

```python
df1=df1[~df1['avg cost (two people)'].isna()]
```

```python
df1.describe()
```

```python
import pandas as pd
from scipy import stats


ratings_yes_online = df1[df1['online_order'] == 'Yes']['rate (out of 5)'].dropna()
ratings_no_online = df1[df1['online_order'] == 'No']['rate (out of 5)'].dropna()

# Perform independent two-sample t-test
t_stat_online, p_value_online = stats.ttest_ind(ratings_yes_online, ratings_no_online)

print(f"T-statistic: {t_stat_online}")
print(f"P-value: {p_value_online}")
if p_value_online < 0.05:
    print("Significant difference between groups.")
else:
    print("No significant difference between groups.")
```

```python

ratings_yes_booking = df1[df1['table booking'] == 'Yes']['rate (out of 5)'].dropna()
ratings_no_booking = df1[df1['table booking'] == 'No']['rate (out of 5)'].dropna()

# Perform independent two-sample t-test
t_stat_booking, p_value_booking = stats.ttest_ind(ratings_yes_booking, ratings_no_booking)

print(f"T-statistic: {t_stat_booking}")
print(f"P-value: {p_value_booking}")
if p_value_booking < 0.05:
    print("Significant difference between groups.")
else:
    print("No significant difference between groups.")
```

```python
# Pearson correlation
corr_coeff, p_value = stats.pearsonr(df1['avg cost (two people)'], df1['rate (out of 5)'])

print(f"Correlation Coefficient: {corr_coeff}")
print(f"P-value: {p_value}")
```
